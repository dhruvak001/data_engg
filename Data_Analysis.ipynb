```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/India-Tourism-Statistics-2019-21.csv')
df.fillna({"% Growth 2021-21/2019-20-Domestic": 0, "% Growth 2021-21/2019-20-Foreign": 0}, inplace=True)

# Remove leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# One-hot encode categorical columns (Circle and Name of the Monument)
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df[['Circle', 'Name of the Monument']]).toarray()
# encoded_features[2]

# Combine encoded features with numeric data
X = np.hstack((encoded_features, df[['Domestic-2019-20', 'Foreign-2019-20']].values))
y_domestic = df['Domestic-2020-21'].values
y_foreign = df['Foreign-2020-21'].values

# Scale features and target values
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_y_domestic = MinMaxScaler()
y_domestic_scaled = scaler_y_domestic.fit_transform(y_domestic.reshape(-1, 1))

scaler_y_foreign = MinMaxScaler()
y_foreign_scaled = scaler_y_foreign.fit_transform(y_foreign.reshape(-1, 1))

# Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_domestic_train, y_domestic_test = train_test_split(X_scaled, y_domestic_scaled, test_size=0.2, random_state=42)
_, _, y_foreign_train, y_foreign_test = train_test_split(X_scaled, y_foreign_scaled, test_size=0.2, random_state=42)

# Define Neural Network model
def create_model(input_dim):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(128, activation='relu', input_dim=input_dim),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(1)  # Single output
    ])
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
    return model

# Train model for Domestic Visitors
model_domestic = create_model(X_train.shape[1])
history_domestic = model_domestic.fit(X_train, y_domestic_train, epochs=50, batch_size=16, validation_split=0.2, verbose=1)

# Train model for Foreign Visitors
model_foreign = create_model(X_train.shape[1])
history_foreign = model_foreign.fit(X_train, y_foreign_train, epochs=50, batch_size=16, validation_split=0.2, verbose=1)

# Predict for test data
y_domestic_pred_scaled = model_domestic.predict(X_test)
y_foreign_pred_scaled = model_foreign.predict(X_test)

# Inverse transform predictions and actual values
y_domestic_pred = scaler_y_domestic.inverse_transform(y_domestic_pred_scaled)
y_domestic_actual = scaler_y_domestic.inverse_transform(y_domestic_test)

y_foreign_pred = scaler_y_foreign.inverse_transform(y_foreign_pred_scaled)
y_foreign_actual = scaler_y_foreign.inverse_transform(y_foreign_test)

# Evaluate performance
def evaluate_model(y_actual, y_pred, visitor_type):
    mae = mean_absolute_error(y_actual, y_pred)
    mse = mean_squared_error(y_actual, y_pred)
    r2 = r2_score(y_actual, y_pred)
    print(f"{visitor_type} Visitors - MAE: {mae:.2f}, MSE: {mse:.2f}, R²: {r2:.2f}")
    return mae, mse, r2

evaluate_model(y_domestic_actual, y_domestic_pred, "Domestic")
evaluate_model(y_foreign_actual, y_foreign_pred, "Foreign")

# Plot results for Domestic Visitors
plt.figure(figsize=(12, 6))
plt.plot(y_domestic_actual, label='Actual Domestic Visitors', marker='o', alpha=0.7)
plt.plot(y_domestic_pred, label='Predicted Domestic Visitors', marker='x', alpha=0.7)
plt.title("Domestic Visitors: Actual vs Predicted in Year 2020-21")
plt.xlabel("Sample Index")
plt.ylabel("Number of Visitors")
plt.legend()
plt.grid()
plt.show()

# Plot results for Foreign Visitors
plt.figure(figsize=(12, 6))
plt.plot(y_foreign_actual, label='Actual Foreign Visitors', marker='o', alpha=0.7)
plt.plot(y_foreign_pred, label='Predicted Foreign Visitors', marker='x', alpha=0.7)
plt.title("Foreign Visitors: Actual vs Predicted in Year 2020-21")
plt.xlabel("Sample Index")
plt.ylabel("Number of Visitors")
plt.legend()
plt.grid()
plt.show()
```
